```python
import pandas as pd
import re
import pickle

pd.options.display.max_columns = 100
```

```python
morning = pickle.load(open('jogr112_close_frame','rb')) #close of day before/ morning
afternoon = pickle.load(open('jogr112_intraday_frame','rb')) #intra day
```

```python
afternoon.head()
```

```python
merged = pd.merge(afternoon,morning,left_on=afternoon.index, right_on=morning.index, suffixes=('_afternoon','_morning'))
merged = merged.replace('NA',0)

merged
```

```python
Cash_Bids_Change1 = merged['Cash_Bids_1_afternoon'].astype('float') - merged['Cash_Bids_1_morning'].astype('float')
Cash_Bids_Change2 = merged['Cash_Bids_2_afternoon'].astype('float') - merged['Cash_Bids_2_morning'].astype('float')

Basis_Change1 = merged['Basis1_1_afternoon'].astype('float') - merged['Basis1_1_morning'].astype('float')
Basis_Change2 =merged['Basis1_2_afternoon'].astype('float') - merged['Basis1_2_morning'].astype('float')

```

```python
final = merged[['Month_afternoon', 'Crop_afternoon','Contract_afternoon']]
final['Cash_Bids_Change1'] = Cash_Bids_Change1
final['Cash_Bids_Change2'] = Cash_Bids_Change2
final['Basis_Change1'] = Basis_Change1
final['Basis_Change2'] = Basis_Change2
```

```python
final = final.fillna(0)
```

```python
final
```

```python
out = pd.DataFrame({"Month": final['Month_afternoon'],"Crop": final["Crop_afternoon"],'Cash_bids_difference1':Cash_Bids_Change1,
              "Cash_bids_difference2": Cash_Bids_Change2,
              "Basis_difference1": Basis_Change1,
              "Basis_difference2":Basis_Change2})
out
```
